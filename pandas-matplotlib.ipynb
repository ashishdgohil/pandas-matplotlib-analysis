### FIFA DATA

```python
# loading data
fifa = pd.read_csv('fifa_data.csv')
```

### Histogram chart

```python
# creating bins like 10, 20, 30, 40 etc
bins = [40,50,60,70,80,90,100]

# increasing the chart size
plt.figure(figsize=(8,5))

# plotting histogram on overall skill value & changing chart color
plt.hist(fifa.Overall, bins=bins, color='#E0245E')

# plotting x axis
plt.xticks(bins)

# adding xlable, ylabel & title to the chart
plt.xlabel('Skill Level', fontdict={'fontweight':'bold','fontsize':16})
plt.ylabel('Number of players', fontdict={'fontweight':'bold','fontsize':16})
plt.title('Distribution of player skills in FIFA 2018', fontdict={'fontweight':'bold','fontsize':22})

# command to show the chart down below
plt.show()
```

### Pie chart - 1

```python
# checking in the data in coloumn Preferred Food which players are Left footed with loc(lookup) method 
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

# increasing the chart size
plt.figure(figsize=(8,5))

# adding labels
labels = ['Left', 'Right']

# changing colours
colors = ['#846CCA', '#1CD760']

# pie chart plotting, assigning the label names, assigning colors, adding percentage no's with autopct
# 2f means 2 decimal points after the value, %% means to the percentage sign in the pie chart
plt.pie([left,right], labels=labels, colors=colors, autopct='%.2f %%')

# adding title to chart
plt.title('Foot preference of FIFA players', fontdict={'fontweight':'bold','fontsize':22})

# command to show the chart down below
plt.show()
```

### Pie chart - 2

```python
# increasing the chart size
plt.figure(figsize=(12,5))

# adding a different style 
plt.style.use('ggplot')

# we had to removed LBS value from weight hence we used strip method and as there were few empty spaces in data
# we used if else to check if x is string remove lbs or let x be x as it is and last with int we converted numbers from str to int
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

# checking how many players are below 125 lbs with count method
light = fifa.loc[fifa.Weight < 125].count()[0]

# checking how many players are greater then 125 and less then 150 in weight with AND condition
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]

# checking how many players are greater then 150 and less then 200 in weight with AND condition
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]

# checking how many players are greater then 175 and less then 200 in weight with AND condition
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]

# checking how many players are greater then 200 in weight with AND condition
heavy = fifa.loc[fifa.Weight >= 200].count()[0]

# assigning all the values created above to a single value ie weights
weights = [light, light_medium, medium, medium_heavy, heavy]

# adding labels
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']

# create explode property - which will create partition / split inside the pie chart
explode = (.4,.2,0,0,.4)

# adding title
plt.title('Weight distribution of FIFA players (in lbs)', fontdict={'fontweight':'bold','fontsize':22})

# plotting pie chart, assigning labels to labels value created above, adding pecentage symbol and no in chart, adding the distance between percentages if they are overlapping
plt.pie(weights, labels=labels,  autopct='%.2f %%', pctdistance=0.9, explode=explode)

# command to show the chart down below
plt.show()
```

### Box & Whisker chart

```python
# changing style for chart
plt.style.use('default')

# resizeing the chart size
plt.figure(figsize=(5,8))

# checking the overall values index no of players play for the club below
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
manutd = fifa.loc[fifa.Club == 'Manchester United']['Overall']
juve = fifa.loc[fifa.Club == 'Juventus']['Overall']

# adding labels
labels = ['Barcelona', 'Real Madrid', 'Man Utd', 'Juventus']

# adding x & y axis labels
plt.ylabel('FIFA overall rating')

# adding title
plt.title('European football teams comparision',fontdict={'fontweight':'bold','fontsize':22})

# plotting box & whisker chart, assigning labels to labels created above, patch artist inserted blue color
plt.boxplot([barcelona,madrid,manutd,juve], labels=labels, patch_artist=True, medianprops={'linewidth':2})

# command to show the chart down below
plt.show()
```
